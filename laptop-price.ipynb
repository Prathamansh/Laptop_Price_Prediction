```python
def fetch_processor_name(text):
    if "Core" in text:
        return text
    elif text.split()[0]=='Intel':
        return " ".join(text.split()[0:2])
    elif text.split()[0]=='AMD':
        if text[4]=='A':
            return "AMD A-Series"
        elif text[4]=='E':
            return "AMD E-Series"
        elif text[4]=='R':
            return "AMD Ryzen Series"
        elif text[4]=='F':
            return "AMD FX Series"
    else:
        return "Another processors"
```

```python
x_train.sample()
```

```python
step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= LinearRegression()
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
# some features are which are irrelvent or noisy
# features are highly correalted(multicollinearinty)
# means if i has apple then my os will be ios / macos for sure
```

```python
#another problem is:
# too many features
# overfitting
```

```python
# # two type of regularizations:
# 1. L2 Regularization(Ridge Regression): Penalize the large coefficient,  the model still use all the features but keep their weights small
# generally works when there are many small features / medium useful features
#2. L1 Regularization (Lasso Regression)
#          stricter than L2, it can actually set some coefficient to the 0, if the feature isnt helping enough
```

```python
# Ridge regression

step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= Ridge()#having no impact
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
# Lasso Regression

step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= Lasso()  # there is no such col. which has impact on data as we can change the value of alpha
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
#knn algo regressor

step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= KNeighborsRegressor(n_neighbors=4)
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
# its giving less value of r2 score which is not good
```

```python
# next algo

# svm, support vector regression
step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= SVR(C=100000000)
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
# decision tree
step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= DecisionTreeRegressor(max_depth=9)
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
# Ada boost algo
step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= AdaBoostRegressor(n_estimators=8)
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```

```python
# xgb regression, it has its own memory something so we dont use that

step1= ColumnTransformer(transformers=[('col_tnf',OneHotEncoder(drop='first'),[0,1,2,4,5])], remainder='passthrough')
step2= XGBRegressor(max_depth =4, n_estimators=95)
pipe = Pipeline([('step1', step1),('step2',step2)])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
print("R2 score: ",r2_score(y_test,y_pred))
print("RMSE : ",root_mean_squared_error(y_test, y_pred))
print("MAE: ", mean_absolute_error(y_test, y_pred))
```
